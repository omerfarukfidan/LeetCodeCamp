```
#WRONG SOLUTION FORGET THAT IT IS SHOULD BE SUBARRAY
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:

        nums.sort()
        sub = []
        sum_target = 0

        if abs(target-nums[len(nums)-1]) >= abs(target-nums[0]): #finding the closes edge as value CASE OF LEFT IS CLOSER
            i = 0
            while sum_target < target and i <= len(nums) - 1:
                sum_target += nums[i]
                sub.append(nums[i])
                i += 1
                if sum_target >= target:
                    return len(sub)
                else:
                    return 0
        elif abs(target-nums[len(nums)-1]) <= abs(target-nums[0]): #CASE OF RIGHT IS CLOSER TO THE TARGER
            i = len(nums)-1
            while sum_target < target and i >= 0:
                sum_target += nums[i]
                sub.append(nums[i])
                i -= 1
                if sum_target >= target:
                    return len(sub)
                else:
                        return 0
```

```
#VIDEO SOLUTION TWO POINTER + SLIDING WINDOWS
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        total = 0
        left = 0
        result = float('inf')
        #right - left +1 gives us length of subarray
        for right in range(len(nums)):
            total += nums[right]
            while total >= target:
                result = min(result, right-left+1)

                total -= nums[left]
                left += 1

        if result == float('inf'):
            return 0
        else:
            return result


       
```
